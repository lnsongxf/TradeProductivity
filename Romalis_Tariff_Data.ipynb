## The Competitive Effects of Trade Liberalization in North America

#### An Empirical Application of the Melitz-Ottaviano model

This project aims to estimate the short- and long run effects of trade liberalization in North America by using an estimation technique developed in <a href = "http://www.sciencedirect.com/science/article/pii/S0022199608001049">Chen, Imbs and Scott (2009)</a> that is based on the <a href = "http://scholar.harvard.edu/files/melitz/files/melot_restud_final.pdf">Melitz and Ottaviano (2008)</a> model.
To this extent, the present workbook constructs a panel dataset containing information on the trade barriers and competitive outcomes in nine manufacturing industries in Canada, Mexico and the US. The observation unit is an industry/country-pair combination, of which there will be 27 in total, followed over the years 1981 to 2012, even though at present data limitations keep us from utilizing the full range of years in our estimation procedure.

## Estimating the Model Equations

The first regression we want to run is the one that has the changes in relative price levels as the dependent variable. Recall that the theory predicts that in the short run, a higher number of firms and lower trade barriers lead to lower prices, while in the long run lower trade barriers will be associated with higher prices and a larger market size will lower domestic prices:
$$ \begin{split}
\Delta \ln \left( \frac{\bar{p}_{it}}{\bar{p}_{it}^*} \right) = &\beta_0 + \beta_1 \Delta \ln \tau_{it} + \beta_2 \Delta \ln \tau_{it}^* + \beta_3 \Delta \ln D_{it} + \beta_4 \Delta \ln D_{it}^* \\ &+ \gamma \left[ \ln \left( \frac{\bar{p}_{it-1}}{\bar{p}_{it-1}^*} \right) + \delta_0 + \delta_1 \ln L_{t-1} + \delta_2\ln L_{t-1}^* + \delta_3 \ln  \tau_{i,t-1} + \delta_4 \ln  \tau_{i,t-1}^* \right] + \varepsilon_{ijt} 
\end{split} $$
Translating the theoretical predictions into our regression coefficients, we expect $\beta_1 < 0$, $\beta_2 >0$, $\beta_3 > 0$, $\beta_4 < 0$, and, provided that our error correction model is correctly specified (such that $-1 < \gamma < 0$), in the long run $\delta_1 < 0$, $\delta_2 > 0$, $\delta_3 < 0$, and $\delta_4 > 0$.
